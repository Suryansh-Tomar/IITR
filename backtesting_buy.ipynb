```python
df=pd.read_csv('check_filtered_stocks_6cond2023.csv') #signals for 2023
df2=pd.read_csv('buy_6cond2024.csv') #signals for 2024
```

```python
df['DATE1'] = pd.to_datetime(df['DATE1'])

# Extract distinct dates from the 'date' column
dist = df['DATE1']

# Format the dates as "DDMMYYYY"
dates = dist.dt.strftime('%Y-%m-%d').tolist()

# Display the list of formatted dates
print(dates)
print(len(dates))
```

```python
df2['DATE1'] = pd.to_datetime(df2['DATE1'])

# Extract distinct dates from the 'date' column
dist2 = df2['DATE1']

# Format the dates as "DDMMYYYY"
dates1 = dist2.dt.strftime('%Y-%m-%d').tolist()
print(dates1)
print(len(dates1))
```

```python
lst=dates+dates1
dates=lst
print(len(dates))
print(dates)
```

```python
concat_df = pd.concat([df, df2], axis=0)
df=concat_df #combining signals from 2023 and 2024 into a single dataframe
df.reset_index(drop=True, inplace=True)
```

```python
concat_df = pd.concat([data, data2], axis=0)
data=concat_df  #combining entire data of 2023 and 2024 into a single dataframe
data.reset_index(drop=True, inplace=True)
```

```python
df['entry_date']='0000-00-00'
df['entry_price']=0
df['stoploss']=0
df['target']=0
df['exit_date']=" "
df['exit_price']=0
df['status']=" "
df['Profit/Loss']=0
for i in range(0,len(dates)):
    df.at[i,'stoploss']=df["LOW_PRICE"][i]
    j=dates2.index(dates[i])   
    dat=pd.to_datetime(dates2[j+1], format='%Y-%m-%d')
    sym=df["SYMBOL"][i]
    filtered_df=data[(data['SYMBOL'] == sym) & (data["DATE1"]==dat)]
    if not filtered_df.empty:
        open_price = filtered_df["OPEN_PRICE"].values[0]
    if(df['stoploss'][i]>open_price):
        continue
    df.at[i,'entry_price']=open_price
    df.at[i,'entry_date']=dates2[j+1]
    #df.at[i,'stoploss']=df["LOW_PRICE"][i]
    df.at[i,'target']=df["entry_price"][i]+((df["entry_price"][i])*0.1)

    k=j+1
    new_df=data[(data['SYMBOL'] == sym)]
    while(k<len(dates2)):
        date=pd.to_datetime(dates2[k], format='%Y-%m-%d')
        filtered_df=new_df[new_df["DATE1"]==date]
        if not filtered_df.empty:
            if  ( filtered_df["HIGH_PRICE"].max()>=df['target'][i]):
                df.at[i,'exit_date']=dates2[k]
                df.at[i,'exit_price']=df['target'][i]
                df.at[i,'status']='target'
                df.at[i,'Profit/Loss']=df["entry_price"][i]-df["exit_price"][i]
                break
            if filtered_df["CLOSE_PRICE"].max()<=df['stoploss'][i]:
                df.at[i,'exit_date']=dates2[k]
                df.at[i,'exit_price']=filtered_df["CLOSE_PRICE"].max()
                df.at[i,'status']='stoploss'
                df.at[i,'Profit/Loss']=df["entry_price"][i]-df["exit_price"][i]
                break
        k=k+1
    
```
